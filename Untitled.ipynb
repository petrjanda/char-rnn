```python
import pandas as pd
import seaborn as sbd
import matplotlib.pyplot as plt
ewma = pd.stats.moments.ewma
```

```python
df = pd.read_csv("adagrad.log", header=None)
df.columns = ["i", "loss", "grad", "time"]

df3 = pd.read_csv("lookup.log", header=None)
df3.columns = ["i", "loss", "grad", "time"]
```

```python
ewma(df["loss"], span=100).plot(figsize=(16,10),label="optim=adagrad")
ewma(df3["loss"], span=100).plot(figsize=(16,10),label="optim=rmsprop")
plt.legend()
```

```python
ewma(df["grad"], span=50).plot(figsize=(16,5))
ewma(df3["grad"], span=50).plot(figsize=(16,5))
```

```python
df["time"].plot(figsize=(16,5))
df3["time"].plot(figsize=(16,5))
```
